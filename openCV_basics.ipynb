```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

COLOR MAPPING

```
img = cv2.imread('puppy.png')
plt.imshow(img)
```

```
# OpenCV takes image in BGR. We need to fix this back to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```
# From RGB to HSV
# HSV Hue Saturation Value
img_hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
plt.imshow(img_hsv)
```

```
# From RGB to HSL
# HSV Hue Lightness Saturation
img_hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
plt.imshow(img_hls)
```

BLENDING AND PASTING IMAGES

```
img1 = cv2.imread('puppy.png')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread('text.jpg')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```
# addWeighted add equal size images
b_img = cv2.resize(img1, (600,600))
f_img = cv2.resize(img2, (600,600))
blended = cv2.addWeighted(src1=b_img, alpha=0.5,src2=f_img, beta=0.5, gamma=0)
plt.imshow(blended)
```

```
# Pasting one mage over another
large_img = img2
small_img = img1
x_offset = 0
y_offset = 0
x_end = x_offset + small_img.shape[1]
y_end = y_offset + small_img.shape[0]

large_img[y_offset:y_end, x_offset:x_end] = small_img
plt.imshow(large_img)
```
