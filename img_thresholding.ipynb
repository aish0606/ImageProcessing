```
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```
def show_img(img):
  fig = plt.figure(figsize=(10,10))
  ax = fig.add_subplot(111)
  ax.imshow(img, cmap='gray')
```

```
img = cv2.imread('DATA/crossword.jpg', 0)
show_img(img)
```

```
img_rainbow = cv2.imread('DATA/rainbow.jpg', 0)
plt.imshow(img_rainbow, cmap='gray')
```

SIMPLE THRESHOLDING

```
ret, img_rain_thres = cv2.threshold(img_rainbow, 127, 255, cv2.THRESH_BINARY)
plt.imshow(img_rain_thres, cmap='gray')
```

```
# dst(x,y)={threshold if src(x,y)>thresh otherwise src(x,y)
ret,thresh3 = cv2.threshold(img_rainbow,127,255,cv2.THRESH_TRUNC)
plt.imshow(thresh3, cmap='gray')
```

```
ret, threshold_img1 = cv2.threshold(img, 180, 255, cv2.THRESH_BINARY)
show_img(threshold_img1)
```

```
ret, threshold_img2 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)
show_img(threshold_img2)
```

ADAPTIVE THRESHOLD

```
adapt_thres = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 15, 8)
show_img(adapt_thres)
```

```
blended = cv2.addWeighted(src1=threshold_img1, alpha=0.6, src2=adapt_thres, beta=0.4, gamma=0)
show_img(blended)
```

```
thres4 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 8)
show_img(thres4)
```

OTSU THRESHOLDING

```
ret, otsu = cv2.threshold(img, 0, 255, cv2.THRESH_OTSU)
print(ret)
show_img(otsu)
```
